```python
import numpy as np
import matplotlib.pyplot as plt

# models and interpolation methods
models = ['Linear Regression', 'Ridge Regression', 'Random Forest', 'XGBoost', 'K Nearest Neighbors']
interpolations = ['Linear', 'Quadratic', 'Spline']

# mean RMSE for each interpolation method
mean_rmse = {
    'Linear': [1.983923, 1.981964, 1.437616, 1.196074, 1.976158], 
    'Quadratic': [1.989851, 1.987848, 1.462203, 1.191008, 1.957310], 
    'Spline': [2.103389, 2.101747, 1.512897, 1.295102, 2.045823],  
}

# stdev for each method
stdev_rmse = {
    'Linear': [0.046861, 0.047337, 0.024294, 0.013575, 0.032711],   
    'Quadratic': [0.051151, 0.051396, 0.040483, 0.013908, 0.025330],  
    'Spline': [0.068411, 0.070273, 0.024648, 0.016587, 0.037261], 
}

fig, ax = plt.subplots(figsize=(10, 6))

# set positions of bars on x axis
x = np.arange(len(models)) 
width = 0.25  

ax.bar(x - width, mean_rmse['Linear'], width, yerr=stdev_rmse['Linear'], label='Linear', capsize=5)
ax.bar(x, mean_rmse['Quadratic'], width, yerr=stdev_rmse['Quadratic'], label='Quadratic', capsize=5)
ax.bar(x + width, mean_rmse['Spline'], width, yerr=stdev_rmse['Spline'], label='Spline', capsize=5)

ax.set_ylabel('Average Test RMSE')
ax.set_title('Comparison of Models Across Different Interpolation Methods')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=30, ha='right')
ax.legend(title='Interpolation Method')

plt.tight_layout()
plt.savefig('interp_comp.png', dpi = 300) 
plt.show()
```
