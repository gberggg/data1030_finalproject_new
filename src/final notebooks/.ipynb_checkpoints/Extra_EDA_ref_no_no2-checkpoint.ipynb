```python
import pandas as pd
from datetime import datetime,timedelta,timezone
import pytz
import smtplib
import numpy as np
import urllib.request
from pandas import json_normalize
import requests
import json
```

```python
# EXTRA EDA: LOOKING AT REFERENCE NO AND NO2

# NO2 for 2023
url_2 = "https://aqs.epa.gov/data/api/sampleData/bySite?email=grace_berg@brown.edu&key= sandkit51&param=42602&bdate=20230101&edate=20231231&state=44&county=007&site=1010"  
response_2 = requests.get(url_2)

# parse out the json file
ref_no2 = response_2.json() 
ref_no2 = json_normalize(ref_no2, record_path=['Data'], meta=['Header'])

ref_no2_df = pd.DataFrame(ref_no2)


# NO for 2023
url_3 = "https://aqs.epa.gov/data/api/sampleData/bySite?email=grace_berg@brown.edu&key= sandkit51&param=42601&bdate=20230101&edate=20231231&state=44&county=007&site=1010"  
response_2 = requests.get(url_2)

# parse out the json file
ref_no = response_2.json() 
ref_no = json_normalize(ref_no, record_path=['Data'], meta=['Header'])

ref_no_df = pd.DataFrame(ref_no)
```

```python
# clean up columns and filter for just FRM
ref_no2_FRM = ref_no2_df[ref_no2_df["method_type"] == "FRM"]
ref_no2_FRM = ref_no2_FRM[["date_gmt","time_gmt","sample_measurement"]]

ref_no_df = ref_no_df[["date_gmt","time_gmt","sample_measurement"]]
```

```python
print(ref_no_df.head())
```

```python
# fix up dates
ref_no2_FRM["datetime_gmt"] = pd.to_datetime(ref_no2_FRM["date_gmt"]+" "+ref_no2_FRM["time_gmt"], format = "%Y-%m-%d %H:%M",utc = True)
ref_no2_FRM['date'] = ref_no2_FRM['datetime_gmt'].dt.tz_convert('US/Eastern')

ref_no_df["datetime_gmt"] = pd.to_datetime(ref_no_df["date_gmt"]+" "+ref_no_df["time_gmt"], format = "%Y-%m-%d %H:%M",utc = True)
ref_no_df['date'] = ref_no_df['datetime_gmt'].dt.tz_convert('US/Eastern')


# clean up columns
ref_no2_FRM = ref_no2_FRM[["date","sample_measurement"]]
ref_no_df = ref_no_df[["date","sample_measurement"]]
```

```python
# rename columns
ref_no2_FRM.rename(columns={'sample_measurement': 'no2_ref'}, inplace=True)
ref_no_df.rename(columns={'sample_measurement': 'no_ref'}, inplace=True)
```

```python
print(ref_no_df.head())
```

```python
# merge
merge_ref_df = ref_no2_FRM.merge(ref_no_df,how='outer',on='date')
```

```python
print(merge_ref_df.head())
```

```python
import matplotlib
from matplotlib import pylab as plt

merge_ref_df.plot.scatter('no2_ref','no_ref', alpha=0.6,s=10) # alpha=0.1,s=10
plt.axline((0, 0), slope=1, color='red', linestyle='--', label='1:1 line')
plt.xlabel('NO2')
plt.xticks(fontsize=9)
plt.ylabel('NO')
plt.title('')
plt.savefig('figures/EDA_figures/ref_no2_vs_no.png') 
plt.show() 
print(" ")
```

```python
# Load some LCS data to get temperature 

temp = pd.read_csv("processed_data_1/sensor1.csv")
print(temp.dtypes)
temp = temp[["date","temp"]]

# convert 'date' column to datetime with `utc=True` to handle mixed timezones
temp['date'] = pd.to_datetime(temp['date'], errors='raise', utc=True) 

# convert back to eastern time
temp['date'] = temp['date'].dt.tz_convert('US/Eastern')

print(temp.dtypes)
```

```python
merge_temp_df = merge_ref_df.merge(temp,how = 'inner',on='date')

print(merge_temp_df.head())
```

```python
merge_temp_df.plot.scatter('no2_ref','no_ref', alpha=0.6,s=10,c='temp') # alpha=0.1,s=10
plt.axline((0, 0), slope=1, color='red', linestyle='--', label='1:1 line')
plt.xlabel('NO2')
plt.xticks(fontsize=9)
plt.ylabel('NO')
plt.title('')
plt.savefig('figures/EDA_figures/ref_no2_vs_no_coloredby_temp.png') 
plt.show() 
print(" ")
```
